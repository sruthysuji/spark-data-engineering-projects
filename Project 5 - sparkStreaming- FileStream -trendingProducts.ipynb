```python
#initialise new spark streaming session 
from pyspark import SparkConf,SparkContext
from pyspark.streaming import StreamingContext
```

```python
sc= SparkContext("local[2]","sparkstreaming")

#initialise the streaming context with batch interval (10 seconds)
ssc=StreamingContext(sc,10)
#checkpoint is used for fault tolerance in spark streaming(only for dstream this is applicable not for DF)
#it saves meta data and state of streaming
ssc.checkpoint("/checkpoint_trending_product")

lines = ssc.textFileStream("/transaction")

#splitting the words with ',' delimiter
transaction =  lines.map(lambda line:line.split(","))
products = transaction.filter(lambda fields : len(fields)>1 and fields[5]!='products').map(lambda fields:fields[5])

#map each words to value 1
product_pairs = products.map(lambda product : (product,1))

#reducer - aggregate all values to get product count
product_count = product_pairs.reduceByKey(lambda x,y:x+y)

#transform() - method is applied on Dstream to transform the word values and arrange then in descending order
sorted_products = product_count.transform(lambda x: x.sortBy(lambda val : val[1],ascending =False))

sorted_products.pprint()

ssc.start()
ssc.awaitTermination()
```

```python
ssc.stop()
```
