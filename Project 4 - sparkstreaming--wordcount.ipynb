```python
#initialise new spark streaming session 
from pyspark import SparkConf,SparkContext
from pyspark.streaming import StreamingContext
```

```python
sc= SparkContext("local[2]","sparkstreaming")

#initialise the streaming context with batch interval (5 seconds)
ssc=StreamingContext(sc,5)

#create a Dstream which will connect to URL with netcat and port number (say 9990)
lines = ssc.socketTextStream("localhost",9990)

#split the lines from Dstream to words 
words = lines.flatMap(lambda line : line.split(" "))

#map each words to 1
pairs_dstream = words.map(lambda word : (word,1))

#reducer
wordcount = pairs_dstream.reduceByKey(lambda x,y :x+y)

#pprint() is the action command of dstream
wordcount.pprint()

ssc.start()
ssc.awaitTermination()
```

```python
ssc.stop()
```
